```python
import pymongo
```

```python
#package which enables us to connect to the MongoDB
dbConn = pymongo.MongoClient("mongodb://localhost:27017/") #connecting to the locally running MongoDB Instance
```

```python
dbname = 'Titanic'
db = dbConn[dbname] #connecting to the database named Titanic present in the mongoDB
#if the database is not present then it will create one
```

```python
#show all the databases
print(dbConn.list_database_names())
```

```python
#check if the database exists in database
dblist = dbConn.list_database_names()
print(dblist)
db_name = 'Titanic'
if db_name in dblist:
    print('Database exists')
else:
    print('database does not exist')
```

```python
#connecting to a collection -- connections are analogous to tables
collection_name = 'Titanic_train'
collection = db[collection_name] #connecting to the collection itself present in the database.
```

```python
#checking if the collection exists
if collection_name in db.list_collection_names():
    print('Collection Exists')
else:
    print('No, Collection does not exist')
```

```python
#retrieving all the data from collection
result = collection.find({})
result[3] #printing the third element
```

Whenever we are calling the find() method, if we have too many rows/records in the table, then we will get too many responses for that which might overwhelm the application. So to avoid this, we have a way in mongodb to have an upper limit on the number of rows returned.<br>
It is called the <b>limit()</b> method.

```python
#printing only the five rows
result_total = collection.find({}).limit(5)
for res in result_total:
    print(res)
```

```python
#retrieving some columns
result_cols = collection.find({},{'Name','Survived'}) #retrieving two columns
#the scond parameter in find() specifies which column to choose 
for res in result_cols:
    print(res)
```

```python
my_db_query = {'Age':'20'}
#printing all the rows where Age is greater than 20
result_Age = collection.find(my_db_query)
for res in result_Age:
    print(res)
```

```python
my_db_query = {'Age':{'$gt':'80'}}
#printing all the rows where Age is greater than 80
result_Age = collection.find(my_db_query)
for res in result_Age:
    print(res)
```

```python
#sorting in ascending order
my_db_query = {'Age':{'$gt':'80'}}
#printing all the rows where Age is greater than 80
result_Age = collection.find(my_db_query).sort('Name')
for res in result_Age:
    print(res)
```

```python
#sorting in descending order
my_db_query = {'Age':{'$gt':'80'}}
#printing all the rows where Age is greater than 80
result_Age = collection.find(my_db_query).sort('Name',-1)
for res in result_Age:
    print(res)
```

```python
#Deleting records from mongodb
#finding the rows satisfying the given criteria
my_db_query = {'Name':'Joshi, Ms. Bhavya'}
x = collection.delete_one(my_db_query) #the deletion step
print(x.deleted_count)
```

We have successfully deleted on row/record from the collection

Similarly, for deleting multiple rows, we can use <b>collection.delete_many(my_db_query)</b>

For deleting all documents of a collection, we can use <b>collection.delete_many({})</b>

```python
#update values present in collection
update_criteria = {'Name':'Gada, Mr. Utsav'}
new_value = {"$set":{'Name':'Shah, Mr. Utsav'}}
x=collection.update_one(update_criteria,new_value)
print(x.modified_count)
```
